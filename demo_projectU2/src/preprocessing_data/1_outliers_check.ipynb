```python
import pandas as pd
import numpy as np
```

```python
def check_outlier(raw_data):
    q1 = np.quantile(raw_data, .25)
    q3 = np.quantile(raw_data, .75)

    iqr = q3 - q1

    low = q1 - 1.5*iqr
    high = q3 + 1.5*iqr

    row_values = []

    for value in raw_data:
        if value<=high and value>=low:
            row_values.append(0)
        else:
            row_values.append(1)
    
    return row_values
```

```python
df_data = pd.read_csv("../../results/1_checked_dataset.csv")
df_data = df_data.drop(columns=["is_null"])
df_data.head()
```

```python
df_data["age_outlier"] = check_outlier(df_data["age"].values)
df_data["bmi_outlier"] = check_outlier(df_data["bmi"].values)
df_data["glucose_outlier"] = check_outlier(df_data["avg_glucose_level"].values)
```
