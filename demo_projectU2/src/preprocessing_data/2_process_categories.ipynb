```python
import pandas as pd
```

```python
df_data = pd.read_csv("../../results/2_removed_outliers.csv")
df_data.head()
```

```python
df_data["gender"].value_counts()
```

```python
df_data["ever_married"].value_counts()
```

```python
df_data["ever_married"] = df_data["ever_married"].replace({"Yes": 1, "No": 0})
df_data["ever_married"].value_counts()
```

```python
matrix_sex = []

for index in df_data.index:
    sex_value = df_data["gender"][index]

    row = [0, 0]

    if sex_value == "Female":
        row[0] = 1
    else:
        row[1] = 1
    
    matrix_sex.append(row)

df_one_hot_gender = pd.DataFrame(data=matrix_sex, columns=["Female", "Male"])
df_one_hot_gender
```

```python
df_data["work_type"].value_counts()
```

```python
matrix_work = []

for index in df_data.index:
    work_value = df_data["work_type"][index]

    row = [0, 0, 0, 0, 0]

    if work_value == "children":
        row[0] = 1
    elif work_value == "Govt_job":
        row[1] = 1

    elif work_value == "Never_worked":
        row[2] = 1
    
    elif work_value == "Private":
        row[3] = 1
    else:
        row[4] = 1

    matrix_work.append(row)

df_work_encoder = pd.DataFrame(
    data=matrix_work, 
    columns=["children", "Govt_job", "Never_worked", "Private", "Self-employed"])

df_work_encoder
```

```python
df_data["smoking_status"].value_counts()
```

```python
matrix_smoking = []

for index in df_data.index:
    work_value = df_data["smoking_status"][index]

    row = [0, 0, 0, 0]

    if work_value == "formerly smoked":
        row[0] = 1
    elif work_value == "never smoked":
        row[1] = 1

    elif work_value == "smokes":
        row[2] = 1    
    else:
        row[3] = 1

    matrix_smoking.append(row)

df_smoker = pd.DataFrame(
    data=matrix_smoking, 
    columns=["formerly smoked", "never smoked", "smokes", "Unknown"])

df_smoker
```

```python
df_data_process = pd.concat([df_data, df_one_hot_gender, df_smoker, df_work_encoder], axis=1)
df_data_process = df_data_process.drop(columns=["gender", "work_type", "smoking_status"])
```

```python
df_data_process["Residence_type"].value_counts()
```

```python
df_data_process["Residence_type"] = df_data_process["Residence_type"].replace(
    {"Urban":1, "Rural":0}
)
```

```python
df_data_process
```

```python
df_data_process.to_csv("../../results/3_data_coded.csv", index=False)
```
