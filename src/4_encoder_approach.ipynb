```python
import pandas as pd
```

```python
df_data = pd.read_csv("../results/3_scaler_demo.csv")
df_data
```

```python
df_data["sex"].value_counts()
```

```python
matrix_data = []

for index in df_data.index:
    sex_value = int(df_data["sex"][index])

    row = [0,0]
    row[sex_value-1] = 1
    matrix_data.append(row)

df_one_hot = pd.DataFrame(data=matrix_data, columns=["sex_F", "sex_M"])
df_one_hot
```

```python
df_concat = pd.concat([df_data, df_one_hot], axis=1)
df_concat
```

```python
df_concat = df_concat.drop(columns=["sex"])
df_concat.to_csv("../results/4_one_hot.csv", index=False)
```
