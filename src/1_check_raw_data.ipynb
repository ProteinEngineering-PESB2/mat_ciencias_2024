```python
import pandas as pd
```

```python
df_data = pd.read_excel("../raw_data/Basal.xlsx")
df_data
```

```python
df_data.shape
```

```python
df_data.drop(columns=['Unnamed: 0', 'Unnamed: 1', 'redcap_repeat_instrument'], inplace=True)
```

```python
df_data.head(5)
```

```python
df_data.columns
```

```python
df_data = df_data[['sex', 'age', 't2m_pre_sdnn', 'bp_systolic', 'bp_diastolic', 'bp_pam', 'bp_pp',
       'weight kg', 'height cm', 'height mt', 'imc']]
```

```python
df_data
```

```python
df_data = df_data[:-1]
df_data
```

```python
df_data.dtypes
```

```python
df_data["check_height"] = (df_data["height cm"] == df_data["height mt"]*100).astype(int)
df_data["check_height"].value_counts()
```

```python
df_data[df_data["check_height"] == 0]
```

```python
df_data["imc_2"] = df_data["weight kg"]/(df_data["height cm"]/100*df_data["height cm"]/100)
df_data = df_data.round(decimals=3)

df_data["check_imc"] = (df_data["imc"] == df_data["imc_2"]).astype(int)
df_data["check_imc"].value_counts()
```

```python
df_data[df_data["check_imc"] == 0]
```

```python
df_data = df_data[df_data["check_height"] == 1]
df_data.shape
```

```python
df_data.to_csv("../results/1_check_raw_data.csv", index=False)
```
