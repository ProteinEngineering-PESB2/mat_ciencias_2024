```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler, MaxAbsScaler, QuantileTransformer
```

```python
df_data = pd.read_csv("../results/2_check_outliers.csv")
df_data
```

```python
df_col_not_work = df_data[["sex", "is_isolated", "is_lof", "response"]]
df_col_to_work =df_data.drop(columns=["sex", "is_isolated", "is_lof", "response"])
df_col_to_work
```

```python
scaler_instance = StandardScaler()
scaler_instance.fit(df_col_to_work.values)
df_scaler = scaler_instance.transform(df_col_to_work.values)
df_scaler
```

```python
scaler_instance = RobustScaler()
scaler_instance.fit(df_col_to_work.values)
df_scaler = scaler_instance.transform(df_col_to_work.values)
df_scaler
```

```python
scaler_instance = MaxAbsScaler()
scaler_instance.fit(df_col_to_work.values)
df_scaler = scaler_instance.transform(df_col_to_work.values)
df_scaler
```

```python
quantile_transformer = QuantileTransformer(output_distribution='normal', random_state=0)

X_trans = quantile_transformer.fit_transform(df_col_to_work)
X_trans
```

```python
df_scaler = pd.DataFrame(data=df_scaler, columns=df_col_to_work.columns)
df_scaler
```

```python
df_scaler = pd.concat([df_scaler, df_col_not_work], axis=1)
df_scaler
```

```python
df_scaler.to_csv("../results/3_scaler_demo.csv", index=False)
```
